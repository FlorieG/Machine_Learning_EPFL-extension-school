## C.4. The distributions of the age group 0-19 years

## C.5. The distributions of all the age group

````python
fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
                         ncols = 3, # sets the number of columns in the subplot grid
                         figsize = (15,4), # sets the width and height of the figure (in inches)
                         sharey=True) # let the plots share the same y-axis

# the left panel
sns.boxenplot(data=df, x='Language', y='0-19 years',
                ax=axes[0])   # draw the plot on the first axes
axes[0].set_title('The 0-19 years proportion into each language')
axes[0].set_xlabel('Language')

# the middle panel
sns.boxenplot(data=df, x='Language', y='20-64 years',
                ax=axes[1])   # draw the plot on the second axes
axes[1].set_title('The 20-64 years proportion into each language')
axes[1].set_xlabel('Language')

# the right panel
sns.boxenplot(data=df, x='Language', y='65 years or over',
                ax=axes[2])   # draw the plot on the third axes
axes[2].set_title('The 65 years or over proportion into each language')
axes[2].set_xlabel('Language')

fig.tight_layout()
plt.show()
````

## C.6. Use of pairplot : Agricultural area vs others area

````python
sns.pairplot(df, x_vars=["Settlement area", "Wooded area", "Unproductive area"], 
                   y_vars=["Agricultural area"], 
                   hue="Elevation",
                   height=4)
plt.show()
````

At first sight the relation between agricultural area and other land area didn't look like linear, half of the scatter plots is full of point. 
- Between Agricultural area and Settlement area most of communes have a low settlement area (0-30 %) distributed into all the agricultural area (0 to 90 %). 
- It's the same between Agricultural area and unproductive area most of the commune have low unproductive area. Unproductive area seems more correlated to elevation than to agricultural area.
- Between Agricultural area and woodded area the diagonal is more dense than the other part of the graph. We can evaluate if there are correlated. 


````python
r = np.corrcoef(df["Agricultural area"], df["Wooded area"])
r[0, 1]
````

````python
r = np.corrcoef(df["Agricultural area"], df["Settlement area"])
r[0, 1]
````

````python
r = np.corrcoef(df["Agricultural area"], df["Unproductive area"])
r[0, 1]
````

The absolute correlation between Agricultural area and wooded area is a bit higher than between Agricultural area and Unproductive area. Trying to model agricultural area in function of wooded area solely by linear regression would not give good predictions.

# C. Probability

### Tasks :
- Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

Valais canton abbreviated as "VS". <br>
This correspond to the proportion of Commune with elevation over 2000 in the Valais canton.

````python
count_Coummune = df[(df['Elevation'] > 2000)].shape[0] # Total of Commune above 2000 m
count_Valais_commune = df[(df['Elevation'] > 2000) & (df['Canton'] == "VS")].shape[0] # Total of Commune above 2000 m in the canton of Valais
prob_result = count_Valais_commune / count_Coummune
prob_result
````

### Tasks :
- Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise.

From numpy an empty array is created with as dimension : Commune length x Canton length  


````python
Canton_matrix = np.zeros((df.shape[0],len(Canton_uniq))) # Canton_uniq comme from the section B.4.
Canton_matrix
````

This array is transform in dataFrame to add index (Communes) and column names (Canton). 

````python
Canton_df = pd.DataFrame(Canton_matrix, index=df.index,columns=Canton_uniq)
````

With a loop across Commune the dataframe is completed.

````python
for i in df.index : 
    Canton_df.loc[i,(Canton_uniq == df['Canton'][i])]=1
````

Then transposed the table to have Canton as index and Commune as columns. 

````python
Canton_df = Canton_df.T
Canton_df
````
