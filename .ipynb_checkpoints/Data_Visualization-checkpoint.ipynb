# The basic of Data Visualization
## Exploratory Data Analysis

- Organize and summarize the data
- Check for missing data
- Identify outliers
- Refine selection of feature variables that will be used for later analysis
- Discover important features and patterns in the data
- Investigate relationships between different variables of the data
- Obtain quick and informative plots of the data

Example known as Anscombe’s Quartet

Note :

In summary: Data visualization is an important component of exploratory data analysis that complements statistical and analytical approaches by providing a different perspective.

## Graphical summaries

- Bar charts

Categorical data

-  Histogram

Quantitative / continuous data
A histogram helps us determine whether the data has several properties such as:
	• unimodal (one peak), bimodal (two peaks), or uniform (no clear peaks)
	• symmetric or asymmetric
	• skewed left or right
	• outliers
	• span, or the difference between the maximum and minimum values

- Pie charts
- Scatter plots

Relationship between twos continuous variables

- Line graphs
Relationship between twos continuous variables, sames with one y for one x

```python
X = np.linspace(-np.pi, np.pi) #The first command defines X to be a numpy array with 50 values ranging from −π to π.
Y = np.sin(X) + np.cos(X)
plt.plot(X, Y)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.show()
```

```python
#We can also make these tick values look a little nicer, by replacing them with the π symbol. 
plt.plot(X, Y, color="red", linewidth=2.0)
plt.xlim(-4, 4)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],  # list of list of xtick locations
    ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"],  # list of labels
)
plt.show()
```

```python
plt.plot(X, Y, color="red", linewidth=2.0, label="$f(x)$")
plt.xlim(-4, 4)
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"],
)
plt.title("The function $f(x)=\sin(x) + \cos(x)$")
plt.xlabel("x")
plt.ylabel("y=f(x)")
plt.legend()
plt.show(); # Finish command with ; to avoid somes text add into the notebook.
```

## A more technical look at plotting

```python
#Implicit way :
#using pyplot directly, cleaner and simpler, good for simple plots
plt.plot([0,1,2,3,4],[0,10,5,30,30], color='magenta', linewidth=3)
plt.show()
```

```python
#Explicit way
#defining a figure and axes objects, calling the method of the axes objects, better for complicated plots
fig, ax = plt.subplots()
ax.plot([0,1,2,3,4],[0,10,5,30,30], color='magenta', linewidth=3)
plt.show()
```

### Multiple Axes

## Example . barplots

```python
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np

countries = ["USA", "GB", "China", "Russia", "Germany"]
bronzes = np.array([38, 17, 26, 19, 15])
silvers = np.array([37, 23, 18, 17, 10])
golds = np.array([46, 27, 26, 19, 17])
total = np.array([121, 67, 70, 55, 42])

x_cor = np.arange(5)

plt.bar(x_cor, total)
plt.show()
```

```python
plt.bar(x_cor, total)
plt.xticks(x_cor, countries)
plt.show()
```

```python
plt.bar(x_cor, total)
plt.xticks(x_cor, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.title("2016 Olympics Medals: Top Countries")
plt.show()
```

```python
plt.bar(x_cor, total, color="green", edgecolor="black")
plt.xticks(x_cor, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.title("2016 Olympics Medals: Top Countries")
plt.show()
```

```python
plt.bar(x_cor, bronzes, color="#CD853F", label="bronze medals")
plt.bar(x_cor, silvers, color="silver", bottom=bronzes, label="silver medals")
plt.bar(x_cor, golds, color="gold", bottom=silvers + bronzes, label="gold medals")
plt.xticks(x_cor, countries)
plt.ylabel("Medals")
plt.xlabel("Countries")
plt.title("2016 Olympics Medals: Top Countries")
plt.legend(loc="upper right")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
from sklearn import datasets

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris["data"], columns=iris["feature_names"]) #Store data as a pd.DataFrame
iris_df["species"] = iris["target"]
# iris_df.head()
species = ["Iris setosa", "Iris versicolor", "Iris virginica"]
colors  = ["red", "orange", "green"]
```

```python
for i in range(0,3): # That's better
    species_df = iris_df[iris_df["species"]==i]
    plt.scatter(species_df["sepal length (cm)"],species_df["petal length (cm)"], color=colors[i], label=species[i])
plt.xlabel("petal length (cm)")
plt.ylabel("sepal length (cm)")
plt.title("Iris dataset: ppetal length vs sepal length")
plt.legend(loc='lower right')
plt.show()
```

### Task 2 :

```python
fig, axes = plt.subplots(nrows =1, ncols = 3, sharex=True, sharey=True,figsize = (12, 3.5))
for i in range(0,3):
    species_df = iris_df[iris_df["species"]==i]
    axes[i].set(title=species[i], ylabel="petal length (cm)", xlabel = "sepal length (cm)")
    axes[i].scatter(species_df["sepal length (cm)"],species_df["petal length (cm)"], alpha=0.5)
plt.show()
```

```python
# Solutions (PS, I prefer what I have done this time)
# 2nd solution using a for loop and the zip function

fig, axes = plt.subplots(nrows = 1,
                         ncols = 3,
                         figsize = (12, 3.5))

for i, ax in zip(range(0, 3), axes.flatten()):

    # Filter data for each species
    species_df = iris_df[iris_df['species'] == i]
    
    # Populate subplots
    ax.scatter(species_df['sepal length (cm)'], species_df['petal length (cm)'], alpha = 0.7)
    
    # Add x/y labels and titles
    ax.set_xlabel('sepal length (cm)')
    ax.set_ylabel('petal length (cm)')
    ax.set_title(species[i])   
    
    # Apply limits to x/y axes
    ax.set_xlim([4.0,8.2])
    ax.set_ylim([0.5,7.5])

fig.tight_layout()
```

```python
# 3rd solution using a gridspec_kw dictionary

fig, axes = plt.subplots(nrows = 1,  
                         ncols = 3, 
                         figsize = (15, 3.5),
                         gridspec_kw = {"wspace": 0.3,  # Sets the width of padding between subplots (is a fraction of the Axes width)
                                        "width_ratios": [ 1, 2, 1] # Sets relative width ratios between subplots 
                                       }
                        )

for i, ax in zip(range(0, 3), axes.flatten()):

    # Filter data for each species
    species_df = iris_df[iris_df['species'] == i]
    
    # Populate subplots
    ax.scatter(species_df['sepal length (cm)'], species_df['petal length (cm)'], alpha = 0.7)
    
    # Add x/y labels and titles
    ax.set_xlabel('sepal length (cm)')
    ax.set_ylabel('petal length (cm)')
    ax.set_title(species[i])   
    
    # Apply limits to x/y axes
    ax.set_xlim([4.0,8.2])
    ax.set_ylim([0.5,7.5])
    
# fig.tight_layout() # Not necessary with gridspec_kw
```

## Introduction to Seaborn
So far, we have only considered plotting with matplotlib, but in fact, Python offers several alternatives such as:

- Seaborn
- Bokeh
- Pandas
- ggplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset("iris") # way esiaer to downlod this data set
iris.head()
```

```python
sns.histplot(iris["sepal_length"], kde=True, stat="density")
plt.show()
```

```python
sns.regplot(x=iris.sepal_length, y=iris.petal_length)
plt.title("A Title") # Add plot title
plt.ylabel("y-label") # Adjust the label of the y-axis
plt.xlabel("x-label") # Adjust the label of the x-axis
plt.ylim(0, 10) # Adjust the limits of the y-axis
plt.xlim(0, 10) # Adjust the limits of the x-axis
sns.set_style('dark') #'ticks' -> default
plt.show()
```

```python
sns.pairplot(iris, height=1.5)
sns.set_style('ticks')
plt.show()
```

```python
sns.pairplot(iris, hue="species", corner=True, height=1.5)

plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_length", "sepal_width"], 
                   y_vars=["petal_length"], 
                   hue="species", 
                   height=3)
plt.show()
```

```python
#Below is an example of how to combine the Seaborn plots with the subplots method.

fig, axes = plt.subplots(nrows = 1, # sets the number of rows in the subplot grid
                         ncols = 2, # sets the number of columns in the subplot grid
                         figsize = (8, 3), # sets the width and height of the figure (in inches)
                         sharey=True) # let the two plots share the same y-axis

# the left-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue='species', 
                ax=axes[0])   # draw the plot on the first axes


axes[0].set_title('petal length vs sepal length')
axes[0].legend(loc='upper left') # set the legend location
axes[0].set_xlabel('Sepal Length')

# the right-hand side scatter plot
sns.scatterplot(data=iris, x="sepal_width", y="petal_length", hue='species', 
                ax=axes[1],   # draw the plot on the second axes
                legend=False) # turn of the legend
    
axes[1].set_title('petal length vs sepal width')
axes[1].set_xlabel('Sepal Width')

fig.tight_layout()
plt.show()
```

```python
sns.pairplot(
    iris,
    x_vars=["sepal_length"],
    y_vars=["petal_length"],
    height=3.5,
    hue="species",
    palette="pastel",
)
plt.show()
```

```python
# define a custom palette
my_palette = ["#95a5a6", "#e74c3c", "#34495e"]
sns.pairplot(
    iris,
    x_vars=["sepal_length"],
    y_vars=["petal_length"],
    height=3.5,
    hue="species",
    palette=my_palette,   # <-- see here, custom palette
)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
grid = sns.FacetGrid(tips, col="time")   # define the grid and the conditioning variable time into columns
grid = grid.map(sns.scatterplot, "total_bill", "tip") # define the scatter plot
```

```python
grid = sns.FacetGrid(tips, col="time",   # define the conditioning variable
                           hue= 'sex')   # define the color-code variable
grid = grid.map(sns.scatterplot, "total_bill", "tip") # define the scatter plot

grid.add_legend()
plt.show()
```

```python
grid = sns.FacetGrid(tips, col="time",   # define the 1st conditioning variable
                           row= 'sex')   # define the 2nd conditioning variable
grid = grid.map(sns.scatterplot, "total_bill", "tip") # define the scatter plot
```

```python
grid = sns.FacetGrid(tips, col="time",      # define the 1st conditioning variable
                           row= 'sex',      # define the 2nd conditioning variable
                           hue= 'smoker')   # define the color-code variable
grid = grid.map(sns.scatterplot, "total_bill", "tip") # define the scatter plot

grid.add_legend()
plt.show()
```

```python
grid = sns.FacetGrid(tips, col="time", row= 'sex')
```

```python
grid = sns.FacetGrid(tips, col="time", row= 'sex')
grid = grid.map(plt.scatter, "total_bill", "tip") # replace `sns.scatterplot` by `plt.scatter`
```
