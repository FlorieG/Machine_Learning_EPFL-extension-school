````python
class Pet(object):
    def __init__(self, name, animal_type):
        self.name = name
        self.animal_type = animal_type
````

````python
Dexter = Pet("Dexter", "dog")
Dexter.animal_type
````

````python
class Dog(Pet):
    def __init__(self, name, breed):
        Pet.__init__(self, name, "dog")
        self.breed = breed

Ornit = Dog("Ornit", "Vizsla")
Ornit.animal_type
````
